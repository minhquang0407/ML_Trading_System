# Bayes Logistic Regression

Kết quả Net Profit: 55.97% và Max Drawdown: -56.96% là một bản án hoàn toàn hợp lý về mặt khoa học.

- Giới hạn của Toán học Tuyến tính: Bayesian Softmax Regression bản chất vẫn là một Mô hình Tuyến tính (Linear Model).

- Ranh giới quyết định của nó chỉ là một mặt phẳng siêu phẳng: $a_k = \mathbf{w}_k^T \mathbf{\phi}$.Thực tế phi tuyến của thị trường: Thị trường tài chính bị chi phối bởi hàng triệu biến số nhiễu loạn, tâm lý đám đông, và các điểm đứt gãy (như tin tức vĩ mô).

- Dữ liệu này có dạng phi tuyến tính cực kỳ phức tạp.Kết cục (Underfitting): Vì bộ não của V1.0 quá "cứng nhắc" (tuyến tính), nó không thể uốn lượn để phân tách các điểm dữ liệu phức tạp.

- Để giảm thiểu hàm Loss một cách lười biếng nhất, nó quyết định dự đoán theo xu hướng chính của tập Train (Đoán toàn thị trường là Tăng - Lớp 1). Nó đã hoàn toàn mù lòa trước các tín hiệu Bán Khống (Lớp 3).
